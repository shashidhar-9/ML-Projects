```python


from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
model = load_model('/content/drive/My Drive/cancer/mymodel.h5')
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
img = image.load_img('/content/drive/My Drive/cancer/cancer/val/Normal/22.jpeg', target_size=(150, 150))  #Normal/22.jpeg
imgplot = plt.imshow(img)
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
images = np.vstack([x])
classes = model.predict_classes(images, batch_size=10)
if classes == [1]:
   print("cancer")
else:
   print("Normal")
```
